## Metabolism of Itraconazole by the Human Microbiota: Identification of Five Potential Drug Metabolites

### Filter matches for ITRACONAZOLE

```python
ITRACONAZOLE_matches = matches.matches_df[matches.matches_df['ParentCompound'] == 'ITRACONAZOLE']
ITRACONAZOLE_matches
```

### Keep only one potential metabolite per match index

```python
ITRACONAZOLE_matches_unique_idx = ITRACONAZOLE_matches.drop_duplicates(subset=['Index'])
ITRACONAZOLE_matches_unique_idx
```

### Draw itracoazole molecule

```python
from rdkit import Chem

ITRACONAZOLE = parent = Chem.MolFromSmiles(ITRACONAZOLE_matches_unique_idx['ParentCompoundSmiles'].values[0])
ITRACONAZOLE
```

### Draw potential metabolism products

```python
from rdkit import Chem
from rdkit.Chem import Draw

molecules = [Chem.MolFromSmiles(smiles) for smiles in ITRACONAZOLE_matches_unique_idx['NewCompoundSmiles']]
match3 = Chem.MolFromSmiles(
    'CCC(C)n1ncn(-c2ccc(N(CCNc3ccc(OC[C@H]4CO[C@](Cn5cncn5)(c5ccc(Cl)cc5Cl)O4)cc3)CC(=O)O)cc2)c1=O')
# remove old match3 and add the new one
molecules.remove(molecules[2])
molecules.insert(2, match3)
molecules.insert(0, ITRACONAZOLE)

# Generate legends (e.g., the SMILES strings)
legends = ['ITRACONAZOLE'] + [f"Match {i + 1}" for i in range(len(molecules) - 1)]

# Customize drawing options
options = Draw.MolDrawOptions()
options.minFontSize = 8  # Minimum font size for the legend
options.legendFontSize = 10  # Font size for the legend
#options.moleculeSize = (600, 600)  # Size of each molecule image

# Create the grid image
img = Draw.MolsToGridImage(
    molecules,
    molsPerRow=3,  # Number of molecules per row
    subImgSize=(600, 200),  # Size of each sub-image
    legends=legends,
    useSVG=False,  # Use PNG format
    returnPNG=False,  # Return the image as a PIL Image
    #options=options
)

# Save the image
img.save('molecules_grid.png')
img.show()
```
