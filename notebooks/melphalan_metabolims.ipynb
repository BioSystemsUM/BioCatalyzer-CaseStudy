# Predicted Melphalan Metabolism

## Load the first round results

## Load the first round matches

```python
matches = pd.read_csv('../results/round1/matches.zip', sep='\t', header=0, compression='zip')
matches
```

## Load Reaction Rules

```python
rules = pd.read_csv('../data/rules.zip', sep='\t', header=0, compression='zip')
rules
```

## Predicted metabolism of the drug Melphalan by reaction rule #47002

```python
from rdkit.Chem import AllChem

rxn2 = AllChem.ReactionFromSmarts(new_compounds[new_compounds['NewCompoundID'] == matches.loc[23520].NewCompoundID].NewReactionSmiles.values[0].split(';')[1])
rxn2
```

```python
new_compounds[new_compounds['NewCompoundID'] == matches.loc[23520].NewCompoundID]
```

```python
rr_47002 = AllChem.ReactionFromSmarts(rules[rules['InternalID'] == 'Rule_47002'].SMARTS.values[0])
rr_47002
```
